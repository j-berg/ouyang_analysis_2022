```r
suppressMessages(
    library(DESeq2))
```

```r
input_data <- file.path(getwd(), "output_tables", "counts_compiled.txt")
input_metadata <- file.path(getwd(), "deseq_metadata.txt")
```

```r
count_table <- read.table(
    input_data, 
    sep = '\t', 
    header = TRUE, 
    row.names = 1, 
    check.names=F)
```

```r
# Create conditions dataframe
sample_table <- read.table(
    input_metadata, 
    header = TRUE,
    sep = '\t')
```

```r
for (x in unique(sample_table$Condition)) {
    for (y in unique(sample_table$Condition)) {
        if (x != y) {
         
            # Extract sub-metadata 
            sub_table = sample_table[is.element(sample_table$Condition, c(x, y)),]
            samples = sub_table$ID
         
            # Extract sub-dataframe 
            sub_data = count_table[, is.element(colnames(count_table), samples)]
            
            # Set factors 
            sub_table$Condition <- factor(sub_table$Condition, levels = c(y, x))
            
            # Run 
            dds <- DESeqDataSetFromMatrix(
                countData = sub_data,
                colData = sub_table,
                design = ~ Condition)
            dds <- DESeq(dds)
            res <- results(dds)
            resOrdered <- res[order(res$padj),]

            # Write output to new file
            this_output = file.path(getwd(), "deseq_output", paste(x, "vs",  y, "deseq2.tsv", sep="_"))
            write.table(
                as.data.frame(resOrdered), 
                file = this_output, 
                sep = '\t',  
                row.names = TRUE,
                col.names = NA)   
        }
    }
}
```
